Q2. Consider the system of equations

3x + 2y = 5

10x âˆ’ y = 3

Find the unique solution to these equations using matrix calculus in
Numpy array.

```python
import numpy as np

A = np.array([[3, 2],
              [10, -1]])

b = np.array([5, 3])

solution = np.linalg.solve(A, b)

print("Solution (x, y):")
print(solution)
```

Q4. Generate a 5 by 5 random integer matrix whose values are between 2 &

Q6. Create a Normalized random variable of size 20 using Numpy and find
out each quartile & their corresponding counts.

```python
import numpy as np
random_variable = np.random.normal(loc=0, scale=1, size=20)
print("Normalized Random Variable:")
print(random_variable)
quartiles = np.percentile(random_variable, [25, 50, 75])
print("\nQuartiles:")
print(f"1st Quartile (25th percentile): {quartiles[0]}")
print(f"2nd Quartile (50th percentile, Median): {quartiles[1]}")
print(f"3rd Quartile (75th percentile): {quartiles[2]}")
bins = [-np.inf, quartiles[0], quartiles[1], quartiles[2], np.inf]
counts, _ = np.histogram(random_variable, bins=bins)
print("\nCounts in Each Quartile Range:")
print(f"Below 1st Quartile: {counts[0]}")
print(f"Between 1st and 2nd Quartile: {counts[1]}")
print(f"Between 2nd and 3rd Quartile: {counts[2]}")
print(f"Above 3rd Quartile: {counts[3]}")
```

Q7. Find out the eigen value and eigen vectors of covariance of the data matrix, X=[[4,11],[8,4],[13,5],[7,14]] where each row is one observation while
colums are attribures of that observation. (Hint: Question is related to
PCA & use np.cov and np.linalg.eig functions of numpy package)

```python
import numpy as np
X = np.array([[4, 11],
              [8, 4],
              [13, 5],
              [7, 14]])
cov_matrix = np.cov(X, rowvar=False)
print("Covariance Matrix:")
print(cov_matrix)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

Q9. Draw four line plots using the subplot() function.

Use the following data:

For line 1: x = [0, 1, 2, 3, 4, 5] and y = [0, 100, 200, 300, 400, 500]

For line 2: x = [0, 1, 2, 3, 4, 5] and y = [50, 20, 40, 20, 60, 70]

For line 3: x = [0, 1, 2, 3, 4, 5] and y = [10, 20, 30, 40, 50, 60]

For line 4: x = [0, 1, 2, 3, 4, 5] and y = [200, 350, 250, 550, 450, 150]

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 100, 200, 300, 400, 500]
y2 = [50, 20, 40, 20, 60, 70]
y3 = [10, 20, 30, 40, 50, 60]
y4 = [200, 350, 250, 550, 450, 150]
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y1, marker='o', color='b')
axs[0, 0].set_title('Line Plot 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 1].plot(x, y2, marker='s', color='r')
axs[0, 1].set_title('Line Plot 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[1, 0].plot(x, y3, marker='^', color='g')
axs[1, 0].set_title('Line Plot 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 1].plot(x, y4, marker='d', color='m')
axs[1, 1].set_title('Line Plot 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
plt.tight_layout()
plt.show()
```

Q12. You have a dataset containing sales data for a retail store. Design a
program using pandas to:
(a) Load the dataset into a pandas DataFrame.

(b) Calculate the total sales revenue for each month.

(c) Identify the top-selling product for each month.

(d) Visualize the monthly sales trend using a line plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/retail_sales_dataset.csv')
print(df.head())
df['Date'] = pd.to_datetime(df['Date'])
df['YearMonth'] = df['Date'].dt.to_period('M')
df['Revenue'] = df['Quantity'] * df['Price per Unit']
monthly_revenue = df.groupby('YearMonth')['Revenue'].sum().reset_index()
print(monthly_revenue)
monthly_product_sales = df.groupby(['YearMonth', 'Product Category'])['Quantity'].sum().reset_index()
top_selling_products = monthly_product_sales.loc[monthly_product_sales.groupby('YearMonth')['Quantity'].idxmax()]
print(top_selling_products)
plt.figure(figsize=(10,6))
plt.plot(monthly_revenue['YearMonth'].astype(str), monthly_revenue['Revenue'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


```

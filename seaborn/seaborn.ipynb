```python
import seaborn as sns
import numpy as np
```

```python
# when we consider 1 feature, it is called univariate analysis

# for 2 features it is called bivariate analysis
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
## correlation with heatmap
```

```python
numeric_df = df.select_dtypes(include=[np.number])
numeric_df.corr()
```

```python
sns.heatmap(numeric_df.corr( ))
```

```python
# joinplot is for univariate analysis
```

```python
sns.jointplot(x='tip', y='total_bill', data=numeric_df, kind='reg')
```

```python
# pair plot for more than 2 independent features
#pairplot is also known as scatter plot in which one variable in the same row is matched with the another variable value.
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

```python
#distplot helps to check the distribution of the columns feature
```

```python
sns.displot(df['tip'],kde=False)
```

```python
## categorical plotsssss
#boxplot
#violinplot
#countplot
#bar plot
```

```python
## count plot
sns.countplot(x='day',data=df)
```

```python
#bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```
